#### 1A. Forward Algorithm 

```python
import pandas as pd

x = ['3','5','4','2','1','6','5','4','6','6']

fair_prob = { '1': 1/6,
              '2': 1/6,
              '3': 1/6,
              '4': 1/6,
              '5': 1/6,
              '6': 1/6 }

loaded_prob = { '1': 1/10,
                '2': 1/10,
                '3': 1/10,
                '4': 1/10,
                '5': 1/10,
                '6': 1/2 }

t_prob = {'F': [.95,.095],
          'L': [.9 ,.045]}

forward_score = [[],[]]

begin = [.5,.5]
end = .005

                    #### Forward Algorithm ####

#### Initial condition ####

temp_fair = fair_prob['3']*begin[0]
temp_loaded = loaded_prob['3']*begin[1]

forward_score[0].append(temp_fair)
forward_score[1].append(temp_loaded)

#### Recursion step ####

for i in range(1,len(x)):
    for j in range(len(forward_score)-1):

        temp_fair = fair_prob[x[i]] * (forward_score[0][i-1]*t_prob['F'][0] + forward_score[1][i-1]*t_prob['L'][1])
        temp_loaded = loaded_prob[x[i]] * (forward_score[0][i-1]*t_prob['F'][1] + forward_score[1][i-1]*t_prob['L'][0])

        forward_score[0].append(temp_fair)
        forward_score[1].append(temp_loaded)
    
      
answer = forward_score[0][-1] * end + forward_score[1][-1]*end

forward_data = {'Fair': forward_score[0] ,
               'Loaded': forward_score[1]}

forward_table = pd.DataFrame(forward_data).T

forward_table['End'] = ['-',answer]
forward_table


##### FORWARD ALGORITHM BELOW ####
```

#### 1A. Backward Algorithm

```python
import pandas as pd

x = ['3','5','4','2','1','6','5','4','6','6']

fair_prob = { '1': 1/6,
              '2': 1/6,
              '3': 1/6,
              '4': 1/6,
              '5': 1/6,
              '6': 1/6 }

loaded_prob = { '1': 1/10,
                '2': 1/10,
                '3': 1/10,
                '4': 1/10,
                '5': 1/10,
                '6': 1/2 }

t_prob = {'F': [.95,.095],
          'L': [.9 ,.045]}

back_score = [[],[]]

#begin = .005
end = .5

back_score[0].append(0.05)
back_score[1].append(0.05)

for i in range(1,len(x)):
     for j in range(len(back_score)-1):

            temp_fair = fair_prob[x[-i]] * back_score[0][i-1] * t_prob['F'][0] + loaded_prob[x[-i]] * back_score[0][i-1] * t_prob['L'][1]
            temp_loaded = loaded_prob[x[-i]] * back_score[1][i-1] * t_prob['L'][0] + fair_prob[x[-i]] * back_score[1][i-1] * t_prob['F'][1]
        
            back_score[0].append(temp_fair)
            back_score[1].append(temp_loaded)
            
back_score[0].reverse()
back_score[1].reverse()

backward_data = {'Fair'  : back_score[0] ,
                 'Loaded': back_score[1] }

backward_table = pd.DataFrame(backward_data).T

answer = back_score[0][0] * end * fair_prob[x[0]] + back_score[1][0] * end * loaded_prob[x[0]]

backward_table["End"] = ['-',answer]
backward_table
```

```python
#def transition(seq,l,k):
game_1 = "13523611354216546636464623545431523616256636624125615436156661263416251264661251265142534516464351664651235414612541241463636" 
game_2= "65456364646213523641254216162556366354543152324125615436466561153661263264661251245642314516464351664651265414646363612541266"
game_3 = "46363613523661254216545636464623545431523616255636624125615436143661263466561612264651265142533516464364651661265414612541266"
game_4= "15366166626441543645352166616255636635454315232412241636666121352316125421261264643516651235142525151265414612541266465463636"

seq_1 = "FFFFFFFFFFFFFLLLLLLLLLLLLFFFFFFFFFFLLLLLLLLLLFFFFFFFFFFFLLLLLLLLLFFFFFFLLLLLLLFFFFFFFFFFFFFLLLLLLLLLLLLFFFFFFFFFFFFFFFFLLLLLL" 
seq_2 = "LLLLLLLLLLLLFFFFFFFFFFFFFLLLLLLLLLLFFFFFFFFFFFFFFFFFFFFFFFFFFFLLLLLLLLLLLLLLLLFFFFFFFFFFFFFLLLLLLLLLLLLFFFFFFFFLLLLLLFFFFFFFF"
seq_3 = "LLLLLLFFFFFFFFFFFFFLLLLLLLLLLLLFFFFFFFFFFLLLLLLLLLLFFFFFFFFFFFLLLLLLLLLFFFFFFLLLLLLLFFFFFFFFFFFFFLLLLLLLLLLLLFFFFFFFFFFFFFFFF"
seq_4 = "LLLLLLLLLLLLFFFFFFFFFFFFLLLLLLLLLLLLFFFFFFFFFFFFFFLLLLLLLLLFFFFFFFFFFFFFFFFLLLLLLLLLLLLFFFFFFFFFFFFFFFFFFFFFFFFFFFFFLLLLLLLLL"

seqs = [seq_1,seq_2,seq_3,seq_4]
games = [game_1,game_2,game_3,game_4]

count_FF = [[0],[0],[0],[0]]
count_FL = [[0],[0],[0],[0]]
count_LL = [[0],[0],[0],[0]]
count_LF = [[0],[0],[0],[0]]

prob_FF = []
prob_FL = []
prob_LL = []
prob_LF = []

# Find T(l,k) for each state transition ( F -> F, L->F, L->L, F->L)
#print(count_FF[0][0])

for i in range(len(seqs)):
    for x in range(len(seqs[0])-1):
        if seqs[i][x] == 'F':
            if seqs[i][x] == seqs[i][x+1]:
                count_FF[i][0] += 1
            else: 
                count_FL[i][0] += 1
        else:
            if seqs[i][x] == seqs[i][x+1]:
                count_LL[i][0] += 1
            else:
                count_LF[i][0] += 1

 ## Transition probabilities for state k to k+1               
for i in range(len(count_FF)):
    FF = count_FF[i][0] / (count_FF[i][0] + count_LF[i][0])
    FL = count_FL[i][0] / (count_FL[i][0] + count_LL[i][0])
    LL = count_LL[i][0] / (count_FL[i][0] + count_LL[i][0])
    LF = count_LF[i][0] / (count_LF[i][0] + count_FF[i][0])

    prob_FF.append(FF)
    prob_FL.append(FL)
    prob_LL.append(LL)
    prob_LF.append(LF)

    
### Determine transition probabilities    
max_FF = max(prob_FF)
max_FL = max(prob_FL)
max_LL = max(prob_LL)
max_LF = max(prob_LF)

#print(prob_FF,prob_FL,prob_LL,prob_LF)    

##### Find Ek(b) for each symbol emitted when path pi is in state k #####
# El(1:6) & Ek(1:6)

fcount_1 = [[0],[0],[0],[0]]
fcount_2 = [[0],[0],[0],[0]]
fcount_3 = [[0],[0],[0],[0]]
fcount_4 = [[0],[0],[0],[0]]
fcount_5 = [[0],[0],[0],[0]]
fcount_6 = [[0],[0],[0],[0]]

lcount_1 = [[0],[0],[0],[0]]
lcount_2 = [[0],[0],[0],[0]]
lcount_3 = [[0],[0],[0],[0]]
lcount_4 = [[0],[0],[0],[0]]
lcount_5 = [[0],[0],[0],[0]]
lcount_6 = [[0],[0],[0],[0]]

count_F = [[0],[0],[0],[0]]
count_L = [[0],[0],[0],[0]]

# Number of times the observation occurs given state

for i in range(len(games)):
    for x in range(len(games[0])):
        if seqs[i][x] == 'F':
            if games[i][x] == '1':
                fcount_1[i][0] += 1
            elif games[i][x] == '2':
                fcount_2[i][0] += 1
            elif games[i][x] == '3':
                fcount_3[i][0] += 1
            elif games[i][x] == '4':
                fcount_4[i][0] += 1
            elif games[i][x] == '5':
                fcount_5[i][0] += 1
            elif games[i][x] == '6':
                fcount_6[i][0] += 1
        else:
            if games[i][x] == '1':
                lcount_1[i][0] += 1
            elif games[i][x] == '2':
                lcount_2[i][0] += 1
            elif games[i][x] == '3':
                lcount_3[i][0] += 1
            elif games[i][x] == '4':
                lcount_4[i][0] += 1
            elif games[i][x] == '5':
                lcount_5[i][0] += 1
            elif games[i][x] == '6':
                lcount_6[i][0] += 1
                
prob_f1 = []
prob_f2 = []
prob_f3 = []
prob_f4 = []
prob_f5 = []
prob_f6 = []

prob_l1 = []
prob_l2 = []
prob_l3 = []
prob_l4 = []
prob_l5 = []
prob_l6 = []
 

for i in range(len(fcount_1)):
    f1 = fcount_1[i][0] / seqs[i].count('F')
    l1 = lcount_1[i][0] / seqs[i].count("L")
    f2 = fcount_2[i][0] / seqs[i].count('F')
    l2 = lcount_2[i][0] / seqs[i].count("L")
    f3 = fcount_3[i][0] / seqs[i].count('F')
    l3 = lcount_3[i][0] / seqs[i].count("L")
    f4 = fcount_4[i][0] / seqs[i].count('F')
    l4 = lcount_4[i][0] / seqs[i].count("L")
    f5 = fcount_5[i][0] / seqs[i].count('F')
    l5 = lcount_5[i][0] / seqs[i].count("L")
    f6 = fcount_6[i][0] / seqs[i].count('F')
    l6 = lcount_6[i][0] / seqs[i].count("L")
    
    prob_f1.append(f1)
    prob_l1.append(l1)
    prob_f2.append(f2)
    prob_l2.append(l2)
    prob_f3.append(f3)
    prob_l3.append(l3)
    prob_f4.append(f4)
    prob_l4.append(l4)
    prob_f5.append(f5)
    prob_l5.append(l5)
    prob_f6.append(f6)
    prob_l6.append(l6)

### Determine emission probabilities

max_f1 = max(prob_f1)
max_l1 = max(prob_l1)
max_f2 = max(prob_f2)
max_l2 = max(prob_l2)
max_f3 = max(prob_f3)
max_l3 = max(prob_l3)
max_f4 = max(prob_f4)
max_l4 = max(prob_l4)
max_f5 = max(prob_f5)
max_l5 = max(prob_l5)
max_f6 = max(prob_f6)
max_l6 = max(prob_l6)

print(max_LL, max_LF, max_FF, max_FL, max_f1, max_f2, max_f3, max_f4, max_f5, max_f6, max_l1, max_l2, max_l3, max_l4, max_l5, max_l6)
```
