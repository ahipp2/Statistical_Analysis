```python
from sklearn import preprocessing
import math
import pandas as pd

input_1 = [180,200,8,1]
input_2 = [120,100,4,0]
h1_val = [.000001,.000001,.00001]
h2_val = [.00001,.00005,.00001]
h3_val = [.01,.001,.1]
h4_val = [1,3,.05]

normal_x = [input_1,input_2] 
normalized_x = preprocessing.normalize(normal_x)
print(normalized_x)

### Forward pass - 
net_h1 = normalized_x[0][0]* h1_val[0] + normalized_x[0][1]* h1_val[1] + normalized_x[0][2]* h1_val[2]
net_h2 = normalized_x[0][0]* h2_val[0] + normalized_x[0][1]* h2_val[1] + normalized_x[0][2]* h2_val[2]
net_h3 = normalized_x[0][0]* h3_val[0] + normalized_x[0][1]* h3_val[1] + normalized_x[0][2]* h3_val[2]

out_h1 = 1 / (1 + math.exp(-net_h1) )
out_h2 = 1 / (1 + math.exp(-net_h2) )
out_h3 = 1 / (1 + math.exp(-net_h3) )

net_o1 = out_h1 * h4_val[0] + out_h2 * h4_val[1] + out_h3 *h4_val[2]
out_o1 = 1 / (1 + math.exp(-net_o1))

### Calculated Error
e1_total = .5 * (normalized_x[0][3] - out_o1)**2

### Backward pass - calculate Error total wrt weight
# Error total wrt weight = A * B * C 
# weights right of hidden layer
## A = error total wrt out_o1
A = -(1 - out_o1)
## B = out_o1 wrt net_o1
B = out_o1 * (1- out_o1)
## C = net_o1 wrt w
C = out_h1

error_wrt10 = A*B*C

eta = .05

w_10 = h4_val[0] - eta*error_wrt10
w_11 = h4_val[1] - eta*error_wrt10
w_12 = h4_val[2] - eta*error_wrt10

# weights to left of hidden
# w_1
## D = error total wrt out_h1
D = A * B * w_10
## E = out_h1 wtr net_h1
E = out_h1 * (1-out_h1)
## net_h1 wrt w_1 = I_1
F = normalized_x[0][0]
G = normalized_x[0][1]
H = normalized_x[0][2]

error_wrt1 = D*E*F
error_wrt2 = D*E*G
error_wrt3 = D*E*H
#print(error_wrt1)

w_1 = h1_val[0] - eta * error_wrt1 
w_2 = h1_val[1] - eta * error_wrt1 
w_3 = h1_val[2] - eta * error_wrt1 

w_4 = h2_val[0] - eta * error_wrt2 
w_5 = h2_val[1] - eta * error_wrt2 
w_6 = h2_val[2] - eta * error_wrt2 

w_7 = h3_val[0] - eta * error_wrt3 
w_8 = h3_val[1] - eta * error_wrt3 
w_9 = h3_val[2] - eta * error_wrt3 

#### Patient 2
### Forward pass - 
net_h1 = normalized_x[0][0]* h1_val[0] + normalized_x[0][1]* h1_val[1] + normalized_x[0][2]* h1_val[2]
net_h2 = normalized_x[0][0]* h2_val[0] + normalized_x[0][1]* h2_val[1] + normalized_x[0][2]* h2_val[2]
net_h3 = normalized_x[0][0]* h3_val[0] + normalized_x[0][1]* h3_val[1] + normalized_x[0][2]* h3_val[2]

out_h1 = 1 / (1 + math.exp(-net_h1) )
out_h2 = 1 / (1 + math.exp(-net_h2) )
out_h3 = 1 / (1 + math.exp(-net_h3) )

net_o1 = out_h1 * h4_val[0] + out_h2 * h4_val[1] + out_h3 * h4_val[2]
out_o1 = 1 / (1 + math.exp(-net_o1))

### Calculated Error
e2_total = .5 * (normalized_x[1][3] - out_o1)**2

### Backward pass - calculate Error total wrt weight
# Error total wrt weight = A * B * C 
# weights right of hidden layer
## A = error total wrt out_o1
A = -(0 - out_o1)
## B = out_o1 wrt net_o1
B = out_o1 * (0- out_o1)
## C = net_o1 wrt w
C = out_h1

error_wrt10 = A*B*C

eta = .05

w_10 = h4_val[0] - eta*error_wrt10
w_11 = h4_val[1] - eta*error_wrt10
w_12 = h4_val[2] - eta*error_wrt10

# weights to left of hidden
# w_1
## D = error total wrt out_h1
D = A * B * w_10
## E = out_h1 wtr net_h1
E = out_h1 * (1-out_h1)
## net_h1 wrt w_1 = I_1
F = normalized_x[1][0]
G = normalized_x[1][1]
H = normalized_x[1][2]

error_wrt1 = D*E*F
error_wrt2 = D*E*G
error_wrt3 = D*E*H

error_wrt1 = D*E*F
#print(error_wrt1)

w_1 = h1_val[0] - eta * error_wrt1 
w_2 = h1_val[1] - eta * error_wrt1 
w_3 = h1_val[2] - eta * error_wrt1 

w_4 = h2_val[0] - eta * error_wrt2 
w_5 = h2_val[1] - eta * error_wrt2 
w_6 = h2_val[2] - eta * error_wrt2 

w_7 = h3_val[0] - eta * error_wrt3 
w_8 = h3_val[1] - eta * error_wrt3 
w_9 = h3_val[2] - eta * error_wrt3 

weights = [w_1,w_4,w_7,w_2,w_5,w_8,w_3,w_6,w_9,w_10,w_11,w_12]
pos = ["I1-H1", "I2-H1","I3-H1","I1-H2", "I2-H2","I3-H3","I1-H3", "I2-H3","I3-H3","H1-O", "H2-O", "H3-O"]

print(e1_total)
print(e2_total)
print(out_o1)
error_rates = [e1_total,e2_total,"-","-","-","-","-","-","-","-",'-','-']

data = { "Node:" : pos,
         "Weight" : weights,
         "Error:" : error_rates}
df = pd.DataFrame(data)

df


```
